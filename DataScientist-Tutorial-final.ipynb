```python
import math
#generowanie danych
outF = open("sinus_cosinus.csv","w")
for a in range(-360,360):
    x = a/360.0*math.pi
    outF.write("%f,%f,%f\n" %(x, math.sin(x), math.cos(x)))
outF.close()
```

```python
#https://www.kaggle.com/nitishabharathi/gdp-per-capita-all-countries
import pandas as pd

gdp = pd.read_csv("GDP.csv")
data = gdp[(np.isnan(gdp["1990"])==False) & (np.isnan(gdp["2017"])==False)].sort_values(by="1990")
data
```

```python
plt.plot(data["1990"],data["2017"],'.')
```

```python
np.corrcoef(data["1990"], data["2017"])
```

```python
x = data["1990"]
y = data["2017"]
m,b =np.polyfit(x, y, 1) 

plt.plot(x, y, 'yo', x, m*x+b, '--k') 
plt.show() 
```
